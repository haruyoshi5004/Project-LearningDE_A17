```python
# can actually read DMM?

from pathlib import Path
from struct import unpack, error as structError
from re import findall as reFindall
import os

SCANPETTERN = r"(\d{8})A([0-9.]*)([NS])([0-9.]*)([EW])"

coordinates :list = []

with open(FILEPATH, mode="rb") as file:
    file.seek(100065228, os.SEEK_SET)  # 100065228 はなんか知らんけど固定値で良い
    header = file.read(8).decode()
    print(file.tell())
    if header != "CSMDAT":
        exit()
    print(file.read(34))
    while True:
        if file.tell() > 101000000:
            break
        try:
            chunk = file.read(130)
            # rawCoordinate = file.read(86)
            rawCoordinateByte = chunk[:50]
            rawCoordinate = rawCoordinateByte.decode()
            scanResult = reFindall(SCANPETTERN, rawCoordinate)
            if scanResult == []:
                break
            scanResult = scanResult[0]
            print(chunk)
            print(f"raw: {rawCoordinate}")
            print(scanResult)
            coordinates.append(scanResult)
            # for id in reFindall(SCANPETTERN, rawCoordinate):
            #     print(id)
            file.read(130 * 4) # なんか同じ座標が必ず5つ連続している気がするのでその分をスキップ
        except structError:
            break
```

```python
# generate GPX file from coordinates

import xml.etree.ElementTree as ET
import datetime, os

OUTPUTFILENAME = "tmp/gpxTest.gpx"

root = ET.Element(
    "gpx",
    {
        "creator": "c3p31102@bunkyo.ac.jp",
        "xsi:schemaLocation": "http://www.topografix.com/GPX/1/1 http://www.topografix.com/GPX/1/1/gpx.xsd",
        "version": "1.1",
        "xmlns": "http://www.topografix.com/GPX/1/1",
        "xmlns:xsi": "http://www.w3.org/2001/XMLSchema-instance",
    },
)
trk = ET.SubElement(root, "trk")
trkseg = ET.SubElement(trk, "trkseg")
trkseg.text = "\n"

for coordinate in coordinates:
    # print(coordinate)
    lat_deg = int(coordinate[1][:2])
    lat_minDir = float(coordinate[1][2:])
    lat = lat_deg + lat_minDir / 60
    lon_deg = int(coordinate[3][:3])
    lon_minDir = float(coordinate[3][3:])
    lon = lon_deg + float(lon_minDir) / 60
    # print((lat, lon))
    trkpt = ET.SubElement(trkseg, "trkpt", {
        "lat": f"{lat}",
        "lon": f"{lon}"
    })


ET.ElementTree(root).write(OUTPUTFILENAME, encoding="UTF-8", xml_declaration=True)
```
