```python
# prepare image

import cv2

cv2_img = cv2.imread("tmp/runs/segment/predict/crops/bicycle/IMG_1232.jpg")
# cv2_img = cv2.cvtColor(cv2_img, cv2.COLOR_)
cv2_img = cv2_img.reshape(cv2_img.shape[0] * cv2_img.shape[1], 3)
```

```python
# clustering

from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=5)
cluster.fit(X=cv2_img)

KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,
    n_clusters=5, n_init=10)

cluster.cluster_centers_
```

```python
# output

from IPython.display import display
from PIL import Image as PILIMage

format(255, '02x')
'#ffffff'

cluster_centers_arr = cluster.cluster_centers_.astype(
    int, copy=False)

for rgb_arr in cluster_centers_arr:
    color_hex_str = '#%02x%02x%02x' % tuple(rgb_arr)
    color_img = PILIMage.new(
        mode='RGB', size=(32, 32), color=color_hex_str)
    display(color_img)
```
