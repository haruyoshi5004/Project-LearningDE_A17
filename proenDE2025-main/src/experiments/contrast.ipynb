```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

path = Path("content/public_road/p1_sunny_r4.jpeg")
image = cv2.cvtColor(cv2.imread(str(path.absolute())), cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
alpha = 1.5
beta = 10

adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
plt.imshow(adjusted_image)
```

```python
imageOutput = cv2.cvtColor(adjusted_image, cv2.COLOR_RGB2BGR)
cv2.imwrite("OUTPUT.png", imageOutput)
```
